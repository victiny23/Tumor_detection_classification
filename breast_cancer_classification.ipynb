```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
cancer['data'].shape
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns=np.append(cancer['feature_names'], ['target']))
df_cancer.head()
# 1 for benign, 0 for malignant
```

```python
df_cancer.target.value_counts()
```

```python
df_cancer.describe().transpose()
```

```python
sns.countplot(x='target', data=df_cancer)
```

```python
vis_cor = plt.figure(figsize=(20,12))
df_cancer.corr()['target'][:-1].sort_values().plot(kind='bar')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
abs(df_cancer.corr()['target'][:-1]).sort_values(ascending=False)
```

```python
sns.pairplot(df_cancer, hue='target', vars=['worst concave points', 'worst perimeter', 'mean concave points', 'worst radius', 'mean perimeter'])
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(df_cancer.corr(), annot=True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

```python
X = df_cancer.drop('target', axis=1)
y = df_cancer['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# preprocessing: scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression()
model_log.fit(X_train, y_train)
y_pred_log = model_log.predict(X_test)
cm_log = confusion_matrix(y_test, y_pred_log)
sns.heatmap(cm_log, annot=True)
print(classification_report(y_test, y_pred_log))
```

```python
import matplotlib
matplotlib.__version__
```

```python
# support vector machine
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
y_pred_svc = model_svc.predict(X_test)
cm_svc = confusion_matrix(y_test, y_pred_svc)
sns.heatmap(cm_svc, annot=True)
print(classification_report(y_test, y_pred_svc))
```

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['linear', 'rbf']}
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_svc_grid = grid.predict(X_test)
cm_svc_grid = confusion_matrix(y_test, y_pred_svc_grid)
sns.heatmap(cm_svc_grid, annot=True)
print(classification_report(y_test, y_pred_svc_grid))
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(X_train, y_train)
y_pred_knn = model_knn.predict(X_test)
cm_knn = confusion_matrix(y_test, y_pred_knn)
sns.heatmap(cm_knn, annot=True)
print(classification_report(y_test, y_pred_knn))
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()
model_dtc.fit(X_train, y_train)
y_pred_dtc = model_dtc.predict(X_test)
cm_dtc = confusion_matrix(y_test, y_pred_dtc)
sns.heatmap(cm_dtc, annot=True)
print(classification_report(y_test, y_pred_dtc))
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier
model_rfc = RandomForestClassifier()
model_rfc.fit(X_train, y_train)
y_pred_rfc = model_rfc.predict(X_test)
cm_rfc = confusion_matrix(y_test, y_pred_rfc)
sns.heatmap(cm_rfc, annot=True)
print(classification_report(y_test, y_pred_rfc))
```

```python
# XGBoost
from xgboost import XGBClassifier
model_xgb = XGBClassifier()
model_xgb.fit(X_train, y_train)
y_pred_xgb = model_xgb.predict(X_test)
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
sns.heatmap(cm_xgb, annot=True)
print(classification_report(y_test, y_pred_xgb))
```

```python
# Neural Network
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
X_train.shape
```

```python
model_nn = Sequential()

model_nn.add(Dense(30, activation='relu'))
model_nn.add(Dense(15, activation='relu'))
# binary classification - use sigmoid as last activation function
model_nn.add(Dense(1, activation='sigmoid'))

model_nn.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
print(type(X_train), type(y_train), type(X_test), type(y_test))
```

```python
# convert to the same type 
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

```python
model_nn.fit(x=X_train, y=y_train, epochs=250, validation_data=(X_test, y_test), verbose=False)
```

```python
losses = pd.DataFrame(model_nn.history.history)
losses.head()
```

```python
losses.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)

model_nn2 = Sequential()

model_nn2.add(Dense(30, activation='relu'))
model_nn2.add(Dense(15, activation='relu'))
# binary classification - use sigmoid as last activation function
model_nn2.add(Dense(1, activation='sigmoid'))

model_nn2.compile(loss='binary_crossentropy', optimizer='adam')

model_nn2.fit(x=X_train, y=y_train, epochs=250, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
losses2 = pd.DataFrame(model_nn2.history.history)
losses2.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)

model_nn3 = Sequential()

model_nn3.add(Dense(30, activation='relu'))
model_nn3.add(Dropout(0.5)) # 50% of nureons randomly turned off

model_nn3.add(Dense(15, activation='relu'))
model_nn3.add(Dropout(0.5))

# binary classification - use sigmoid as last activation function
model_nn3.add(Dense(1, activation='sigmoid'))

model_nn3.compile(loss='binary_crossentropy', optimizer='adam')

model_nn3.fit(x=X_train, y=y_train, epochs=250, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
losses3 = pd.DataFrame(model_nn3.history.history)
losses3.plot()
```

```python
y_pred_nn3 = model_nn3.predict_classes(X_test)
cm_nn3 = confusion_matrix(y_test, y_pred_nn3)
sns.heatmap(cm_nn3, annot=True)
print(classification_report(y_test, y_pred_nn3))
```
